## Subtema:  Función de onda Normalizada 

¿Cuál es la amplitud ($A$) de la función de onda $\psi(x)$ al normalizarla? 

Donde: 

$$ \psi(x) = Ae^{-ax^2}$$


$$
\int_{-\infty}^{\infty}  \! \psi(x) \, dx = 1  $$

```python
# Import libraries 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
# Create a linspace  -1 < x < 1 

x = np.linspace(-np.pi,np.pi,102)
```

```python
# Wave function in x-space

A = 1 # Amplitud
a = 1 # constant
psi = A*np.exp(-a*pow(x,2))
psi2 = pow(np.abs(psi),2)
```

```python
# Grafica de la función de onda con A = 1
plt.figure()

plt.plot(x,psi)
plt.plot(x,psi2)

plt.xlim(-np.pi,np.pi)

```

## Función de onda Normalizada 

- $A = (\frac{2a}{\pi})^{1/4}$

```python
# Wave function in x-space

a = 1 # constant
A = pow((2*a/np.pi), 1/4) # Amplitud

psi = A*np.exp(-a*pow(x,2))
psi2 = pow(np.abs(psi),2)
```

```python
# Grafica de la función de onda con A = (2a/pi)^(1/4)

plt.figure()

plt.plot(x,psi, label = '$\psi(x) = Ae^{-ax^2}$')
plt.plot(x,psi2, label = '$|\psi(x)|^2$')
plt.vlines(0,-1,1,linestyle = '--', alpha = 0.3)
plt.text(0,A+0.01, '$A = (\\frac{2a}{\pi})^{1/4} $', fontsize = 12)

plt.xlim(-np.pi,np.pi)
plt.ylim(-0.1,1)
plt.legend()
plt.show()

```
