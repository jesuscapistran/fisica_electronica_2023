## Caracteristicas de ondas y particulas ideales

- Una característica esencial de una partícula es que está localizada en el espacio
- Una onda ideal tiene una sola frecuencia y es infinitamente larga.
- Una onda ideal no se localiza en el espacio

### Objetivo 

- Construir una entidad localizada (paquete de ondas) a partir de ondas infinitamente largas. 

### Paquete de ondas

- Se identifica el paquete de ondas como una partícula porque ¡tiene la naturaleza localizada de una partícula!
- La localización del paquete de ondas corresponde a la posición de la partícula.

### Representación mátematica de onda: 
 - $y_1 = A \cos{(k_1 x- \omega_1 t)}$
 - $y_2 = A \cos{(k_2 x- \omega_2 t)}$
 
 Donde:
 
 - $\omega = 2\pi f$
 - $     k = \frac{2\pi}{\lambda} $
 
 
 ### Principio de superposición
 
 - $y = y1 + y2$
 - $y = A \cos{(k_1 x- \omega_1 t)} + A \cos{(k_2 x- \omega_2 t)}$
 


```python
# Importar librerias (se usaran más adelante en el curso)

import numpy as np  #Esta libreria hace calculo de matrices 
import matplotlib.pyplot as plt  # Esta liberaria nos permite graficar 
```

```python
# Declarar constantes 

kB       = 1.38e-23  # [J/K] Constante de Boltzman 
c        = 2.99e8    # [m/s] Velocidad de la luz en el vacío
T        = 6000      # [K]   Temperatura de la superficie
h        = 6.625e-34 # [J·s] Constante de Planck
q        = 1.602e-19 # [J/eV] 1 joule = 1.602e-19 eV 
```

```python
# Variable independite (tiempo): y(t)
#URL: https://numpy.org/doc/stable/reference/generated/numpy.arange.html

x = np.arange(-6,6,0.0001) # [µm] np.arange(inicio,final,espacio)
x = x*1e-6                 # [m]
x.size
```

```python
## Onda 1
t = 1e-3   # Tiempo de propagación de la onda 
A = 1      # Amplitud de la onda es de 1 
long_01  = 500e-9 # [m] Longitud de onda de 500 nm 
omega_01 = 2*np.pi*(c/long_01)
k_01     = (2*np.pi)/long_01

y1 = A * np.cos(k_01*x - omega_01*t)
```

```python
## Onda 2
t = 1e-3   # La velocidad es de la onda es de 1 ms 
A = 1      # Amplitud de la onda es de 1 
long_02  = 400e-9 # [m] Longitud de onda de 500 nm 
omega_02 = 2*np.pi*(c/long_02)
k_02     = (2*np.pi)/long_02

y2 = A * np.cos(k_02*x - omega_02*t)
```

```python
# Graficar Ondas 1 y 2
# URL: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.vlines.html

plt.figure()
plt.title('Paquete de ondas')

# Graficar curvas
plt.plot(x,y1 ,label = 'y1')
plt.plot(x,y2 ,label = 'y2')
plt.vlines(0,-2,2, linestyles= 'dashed' ,alpha = 0.2)
plt.vlines(0.5e-6,-2,2, linestyles= 'dashed' ,alpha = 0.2)

#plt.plot(x,y1+y2, '-', label = 'y1+y2')

# Muestra las legendas
plt.legend()

# Nombre de los ejes
plt.xlabel('$x [m]$', fontsize=14)
plt.ylabel('Intensidad', fontsize=14)

# Define los limites de las graficas
plt.ylim(-1.5,1.5)
plt.xlim(-1e-6,1e-6)

plt.show()
```

```python
# Graficar Sumda de Ondas 1 y 2
# URL: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.vlines.html

plt.figure()
plt.title('Paquete de ondas: Y1 + Y2')

# Graficar curvas
#plt.plot(x,y1 ,label = 'y1')
#plt.plot(x,y2 ,label = 'y2')

plt.plot(x,y1+y2 ,label = 'y1+y2')
plt.vlines(0,-3.5,3.5, linestyles= 'dashed' ,alpha = 0.2)
plt.vlines(0.5e-6,-3.5,3.5, linestyles= 'dashed' ,alpha = 0.2)
plt.hlines(0,-1,1, alpha = 0.2)

#plt.plot(x,y1+y2, '-', label = 'y1+y2')

# Muestra las legendas
plt.legend()

# Nombre de los ejes
plt.xlabel('$x [m]$', fontsize=14)
plt.ylabel('Intensidad', fontsize=14)

# Define los limites de las graficas
plt.ylim(-3.5,3.5)
plt.xlim(-3e-6,3e-6)

plt.show()
```

## Función envolvente

- Si se combina un gran número de ondas, el resultado es un paquete de ondas que representa una partícula.

$$ y = 2A \cos \left[{\frac{\Delta k}{2}x - \frac{\Delta \omega}{2}t} \right] \cos \left( \frac{k_1 + k_2}{2}x - \frac{\omega_1 + \omega_2}{2}t \right)$$


Donde: 

- $\Delta k = k_1 - k_2$ 
- $\Delta \omega = \omega_1 - \omega_2$ 

Función Primer envolvente: 


$$ 2A \cos \left[{\frac{\Delta k}{2}x - \frac{\Delta \omega}{2}t} \right]$$

```python
# Programa la función de la 1er envolvente 

delta_k     = k_01 - k_02
delta_omega = omega_01 - omega_02 

envolvente = 2*A*np.cos(((delta_k/2)*x - (delta_omega/2)*t))
```

```python
# Graficar Paquete de Ondas  + Envolvente

# Graficar Sumda de Ondas 1 y 2
# URL: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.vlines.html

plt.figure()
plt.title('Paquete de ondas: Y1 + Y2')

# Graficar curvas
#plt.plot(x,y1 ,label = 'y1')
#plt.plot(x,y2 ,label = 'y2')

plt.plot(x,y1+y2,label = 'y1+y2')
plt.plot(x, envolvente, '--r', label = '1er envolvente')
#plt.plot(x, envolvente2, '--r', label = 'envolvente2')


plt.vlines(0,-3.5,3.5, linestyles= 'dashed' ,alpha = 0.2)
plt.vlines(0.5e-6,-3.5,3.5, linestyles= 'dashed' ,alpha = 0.2)
plt.hlines(0,-1,1, alpha = 0.2)

#plt.plot(x,y1+y2, '-', label = 'y1+y2')

# Muestra las legendas
plt.legend()

# Nombre de los ejes
plt.xlabel('$x [m]$', fontsize=14)
plt.ylabel('Intensidad', fontsize=14)

# Define los limites de las graficas
plt.ylim(-3.5,3.5)
plt.xlim(-3e-6,3e-6)

plt.show()
```

La suma de varias ondas de diferentes longitudes de onda ($\lambda$) produce un patrón de interferencia que comienza a localizar la onda. 

## Segunda envolvente (a) 

- Se observa que la primer envolvente: 
    - x= 0, y =  2 
    - x=-2, y = -2 
    - x= 2, y = -2 
- La segunda envolvente debe contener los valores
    - x= 0, y = -2 
    - x=-2, y =  2
    - x= 2, y =  2
    
    
 ### Solución 1
 
- Primer envolvente: 
$$ y = 2A \cos \left[{\frac{\Delta k}{2}x - \frac{\Delta \omega}{2}t} \right] \cos \left( \frac{k_1 + k_2}{2}x - \frac{\omega_1 + \omega_2}{2}t \right)$$

- Segunda envolvente: 
$$ y = -2A \cos \left[{\frac{\Delta k}{2}x - \frac{\Delta \omega}{2}t} \right] \cos \left( \frac{k_1 + k_2}{2}x - \frac{\omega_1 + \omega_2}{2}t \right)$$

```python
# Programa la función de la 1er envolvente 

delta_k     = k_01 - k_02
delta_omega = omega_01 - omega_02 

envolvente = 2*A*np.cos(((delta_k/2)*x - (delta_omega/2)*t))
```

```python
# Programa la función de la 2da envolvente 

delta_k     = k_01 - k_02
delta_omega = omega_01 - omega_02 

envolvente2 = -2*A*np.cos(((delta_k/2)*x - (delta_omega/2)*t))
```

```python
# Graficar Paquete de Ondas  + Envolvente

# Graficar Sumda de Ondas 1 y 2
# URL: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.vlines.html

plt.figure()
plt.title('Paquete de ondas: Y1 + Y2')

# Graficar curvas
#plt.plot(x,y1 ,label = 'y1')
#plt.plot(x,y2 ,label = 'y2')

plt.plot(x,y1+y2, '-k',label = 'y1+y2')
plt.plot(x, envolvente, '--r', label = '1er envolvente')
plt.plot(x, envolvente2, '--b', label = '2da envolvente')


plt.vlines(0,-3.5,3.5, linestyles= 'dashed' ,alpha = 0.2)
plt.vlines(0.5e-6,-3.5,3.5, linestyles= 'dashed' ,alpha = 0.2)
plt.hlines(0,-1,1, alpha = 0.2)

#plt.plot(x,y1+y2, '-', label = 'y1+y2')

# Muestra las legendas
plt.legend()

# Nombre de los ejes
plt.xlabel('$x [m]$', fontsize=14)
plt.ylabel('Intensidad', fontsize=14)

# Define los limites de las graficas
plt.ylim(-3.5,3.5)
plt.xlim(-3e-6,3e-6)

plt.show()
```

## Segunda envolvente (b) 

Se observa que el máximo (cresta) entre la envolvente 1 y le envolvente 2 estan defasadas por 2µm 

- No puede ser un función senoidal, por que sin(0) = 0 
- La función $cos(\theta)$ esta defasada 
    - cos(0) = 1
    - cos (0 + $\pi$) =-1
    - cos (0 + $3\pi$) =-1
    - cos (0 + $5\pi$) =-1
    
    
 - La segunda envolvente tiene valor de -2 , por lo tanto: 


- Primer envolvente: 
$$ A \cos \left[{\frac{\Delta k}{2}x - \frac{\Delta \omega}{2}t} \right] $$

- Segunda envolvente: 
$$ 2A \cos \left[{\frac{\Delta k}{2}x - \frac{\Delta \omega}{2}t} - \pi \right] $$


```python
# Programa la función de la 1er envolvente (Segunda solución)

delta_k     = k_01 - k_02
delta_omega = omega_01 - omega_02 

envolvente = 2*A*np.cos(((delta_k/2)*x - (delta_omega/2)*t))
```

```python
# Programa la función de la 2da envolvente (Segunda Solución)

delta_k     = k_01 - k_02
delta_omega = omega_01 - omega_02 

envolvente2 = 2*A*np.cos(((delta_k/2)*x - (delta_omega/2)*t +np.pi))
```

```python
# Graficar Paquete de Ondas  + Envolvente (Segunda Solución)

# Graficar Sumda de Ondas 1 y 2
# URL: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.vlines.html

plt.figure()
plt.title('Paquete de ondas: Y1 + Y2 (Segunda solución)')

# Graficar curvas
#plt.plot(x,y1 ,label = 'y1')
#plt.plot(x,y2 ,label = 'y2')

plt.plot(x,y1+y2, 'k',label = 'y1+y2')
plt.plot(x, envolvente, 'r', label = '1er Envolvente')
plt.plot(x, envolvente2, 'b', label = '2da Envolvente')


plt.vlines(0,-3.5,3.5, linestyles= 'dashed' ,alpha = 0.2)
plt.vlines(0.5e-6,-3.5,3.5, linestyles= 'dashed' ,alpha = 0.2)
plt.hlines(0,-1,1, alpha = 0.2)

#plt.plot(x,y1+y2, '-', label = 'y1+y2')

# Muestra las legendas
plt.legend()

# Nombre de los ejes
plt.xlabel('$x [m]$', fontsize=14)
plt.ylabel('Intensidad', fontsize=14)

# Define los limites de las graficas
plt.ylim(-3.5,3.5)
plt.xlim(-3e-6,3e-6)

plt.show()
```
