## Ley de Raleigh-Jeans

Para describir la distribución de la energía de un cuerpo negro, resulta útil definir I($\lambda$, T)dl como la intensidad o la potencia por unidad de área emitida en el intervalo de longitud de onda dl. 


El resultado del cálculo **según la teoría clásica** de la radiación de un cuerpo negro, conocido como ley de Rayleigh-Jeans es:

$$ I(\lambda,T) = \frac{2\pi c k_B T}{\lambda^4} $$


Donde: 

- $k_B$     =  Constante de Boltzman
- $ c  $    = [$m/s$] Velocidad de la luz 
- $ T  $    = [K] Temperatura  
- $\lambda$ = [m] Longitud de onda



```python
# Importar librerias (se usaran más adelante en el curso)

import numpy as np  #Esta libreria hace calculo de matrices 
import matplotlib.pyplot as plt  # Esta liberaria nos permite graficar 
```

```python
# Declarar constantes 

kB       = 1.38e-23 # [J/K] Constante de Boltzman  (18012023: Se arregla error de constante)
vel_luz  = 2.99e8  # [m/s] Velocidad de la luz en el vacío
T        = 6000    # [K]   Temperatura de la superficie
```

```python
# Variable independite (Temperatura): P(T)
#URL: https://numpy.org/doc/stable/reference/generated/numpy.arange.html

x = np.arange(0.01,3.01,0.01) # [µm] np.arange(inicio,final,espacio)
Longitud_Onda = x*1e-6     # [m] Longitud de onda en metros

#x 
#Longitud_Onda
```

## Catastrofe ultravioleta (Energía infinita) 

La falta de coincidencia entre teoría y experimentación resultaba tan desconcertante que los científicos le dieron el nombre de catástrofe ultravioleta. (Esta “catástrofe” —energía infinita— se presenta conforme la longitud de onda se aproxima a cero; se le añadió la palabra “ultravioleta” porque las ondas ultravioleta son cortas.)


$$ I(\lambda,T) = \frac{2\pi c k_B T}{\lambda^4} $$

```python
# Programar la ecuación (Varieble dependiente)

Intensidad_6000 = (2*np.pi*vel_luz*kB*6000)/(Longitud_Onda**4)
Intensidad_5000 = (2*np.pi*vel_luz*kB*5000)/(Longitud_Onda**4)
Intensidad_4000 = (2*np.pi*vel_luz*kB*4000)/(Longitud_Onda**4)
Intensidad_3000 = (2*np.pi*vel_luz*kB*3000)/(Longitud_Onda**4)
Intensidad_2000 = (2*np.pi*vel_luz*kB*2000)/(Longitud_Onda**4)

# Revisa las unidades de Intensidad de radiación de cuerpo negro
#Intensidad  # [m/s2][J/K][K]/[m^4] = [J.m/m^2s2]  = [(W/m2)/m^-1]
```

```python
# Graficar Intensidad(Longitud_Onda , T)

plt.figure()
plt.title('Ley de Raleigh-Jeans')

# Graficar curvas
plt.plot(Longitud_Onda, Intensidad_6000, label = '6000 K')
plt.plot(Longitud_Onda, Intensidad_5000, label = '5000 K')
plt.plot(Longitud_Onda, Intensidad_4000, label = '4000 K')
plt.plot(Longitud_Onda, Intensidad_3000, label = '3000 K')
plt.plot(Longitud_Onda, Intensidad_2000, label = '2000 K')

# Muestra las legendas
plt.legend()


# Nombre de los ejes
plt.xlabel('$\lambda [m]$', fontsize=14)
plt.ylabel('Intensidad $ (W/m^2)/m^{-1}$', fontsize=14)


# Define los limites de las graficas
plt.ylim(0,1e15)
plt.xlim(0,1.5e-6)

plt.show()
```
