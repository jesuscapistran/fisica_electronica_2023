```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the lattice constant for silver
a = 0.4086

# Define the lattice vectors
v1 = 0.5 * a * np.array([1, 1, 0])
v2 = 0.5 * a * np.array([1, 0, 1])
v3 = 0.5 * a * np.array([0, 1, 1])

# Define the coordinates of the atoms in the FCC crystal structure for silver
atoms = []
for i in range(4):
    for j in range(4):
        for k in range(4):
            atom = i * v1 + j * v2 + k * v3
            atoms.append(atom)

# Define the coordinates of the corners of the cubic cell
corners = np.array([
    [0, 0, 0],
    a * np.array([1, 0, 0]),
    a * np.array([0, 1, 0]),
    a * np.array([0, 0, 1]),
    a * np.array([1, 1, 0]),
    a * np.array([1, 0, 1]),
    a * np.array([0, 1, 1]),
    a * np.array([1, 1, 1])
])

# Define the edges of the cubic cell
edges = [
    (corners[0], corners[1]),
    (corners[0], corners[2]),
    (corners[0], corners[3]),
    (corners[1], corners[4]),
    (corners[1], corners[5]),
    (corners[2], corners[4]),
    (corners[2], corners[6]),
    (corners[3], corners[5]),
    (corners[3], corners[6]),
    (corners[4], corners[7]),
    (corners[5], corners[7]),
    (corners[6], corners[7]),
]

# Plot the atoms and the cubic cell
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the atoms
for atom in atoms:
    ax.scatter(atom[0], atom[1], atom[2], color='k')

# Plot the cubic cell
for edge in edges:
    ax.plot([edge[0][0], edge[1][0]], [edge[0][1], edge[1][1]], [edge[0][2], edge[1][2]], linestyle='--', color='k')

# Set the axis limits and labels
ax.set_xlim([-0.1 * a, 1.1 * a])
ax.set_ylim([-0.1 * a, 1.1 * a])
ax.set_zlim([-0.1 * a, 1.1 * a])
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

# Show the plot
plt.show()
```
