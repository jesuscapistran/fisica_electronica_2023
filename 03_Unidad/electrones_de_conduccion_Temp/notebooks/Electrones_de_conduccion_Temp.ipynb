```python
import numpy as np
import matplotlib.pyplot as plt

# Definir los valores de las constantes
h_barra = 1.0545718e-34  # [J s]
m = 9.10938356e-31      # masa efectiva de un electrón en un metal típico [kg]

k = 8.617e-5  # Constante de Boltzmann en eV/K
E_F = 4.72  # Energía de Fermi en eV del Litio  a 300 K 
E = np.linspace(0, 10, 1000)  # Energía en eV
T = 1000  # Temperaturas en Kelvin

# Crear una lista de energías (en eV) para graficar
energias = np.linspace(0, 10, 1000)  # rango de energías de 0 a 10 eV

# Calcular la distribución de Fermi-Dirac para cada temperatura
f = 1 / (1 + np.exp((E - E_F)/(k*T)))


# Calcular la densidad de estados para cada valor de energía
densidad_estados = (1 / (2 * np.pi**2)) * ((2 * m) / h_barra**2)**(3/2) * np.sqrt(energias)


Electrones = densidad_estados*f 

# Graficar la densidad de estados
plt.plot(energias, densidad_estados, alpha = 0.2)
plt.plot(energias, Electrones)

plt.xlabel('Energía (eV)')
plt.ylabel('D(E)·f(E) (1/eV)')
plt.title('Densidad de estados en un metal')

# Exportar la figura 
figname = '../fig/Electrones_Conduccon.png'
#plt.savefig(figname, dpi = 300 ,bbox_inches = 'tight')


plt.show()
```
