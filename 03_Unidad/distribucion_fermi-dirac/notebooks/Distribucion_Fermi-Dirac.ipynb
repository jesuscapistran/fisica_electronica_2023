- La ecuación nos dice que la probabilidad de que un estado de energía E esté ocupado por un electrón en un sólido en equilibrio térmico depende de la diferencia entre la energía del estado y la energía de Fermi, así como de la temperatura del sólido.

$$ f(E) = \frac{1}{1 + e^{\frac{E-E_F}{kT}}} $$

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir los parámetros
k = 8.617e-5  # Constante de Boltzmann en eV/K
E_F = 4.72  # Energía de Fermi en eV del Litio  a 300 K 
E = np.linspace(0, 10, 1000)  # Energía en eV

T_values = [0]  # Temperaturas en Kelvin

# Calcular la distribución de Fermi-Dirac para cada temperatura
for T in T_values:
    f = 1 / (1 + np.exp((E - E_F)/(k*T)))

    # Graficar la distribución de Fermi-Dirac
    plt.plot(E, f, label='T = {} K'.format(T))


# Configurar la figura
plt.hlines(0.5, 0, 10, linestyle = 'dashed', alpha = 0.2)
plt.vlines(E_F, -0.2, 1.2, linestyle = 'dashed', alpha = 0.2)
plt.text(0.02,0.52, '$f(E)= 0.5$' )
plt.text(4.8,-0.12, '$E_f(Li)= 4.72$ eV' )

plt.xlim(0,9.5)
plt.ylim(-0.2,1.2)
plt.xlabel('Energía (eV)')
plt.ylabel('Probabilidad de ocupación')
plt.title('Distribución de Fermi-Dirac')
plt.legend()

# Exportar la figura 
figname = '../fig/Fermi_Dirac_0K.png'
#plt.savefig(figname, dpi = 300 ,bbox_inches = 'tight')

plt.show()
```
